```python
df2 = pd.read_csv('D:\Learning\Machine-Learning-\data\dataset_2.csv')
df2.head(5)
```

```python
# Task 1 : In ra index của 10 học sinh có math score và reading score đều nằm trong top 100 
top_10_mathscore_readingscore = df2[(df2['rank_math_score'] <= 100) & (df2['rank_reading_score'] <= 100)] # Combine two condition together 
print(top_10_mathscore_readingscore.index[:10]) # print out the index 
```

```python
df2['rank_reading_score']= df2['reading score'].rank(method='max') 
df2['rank_math_score'] = df2['math score'].rank(method='max')
```

```python
# Task 2: In ra index của 20 học sinh có điểm writing score cao nhất của mỗi group(race/ethnicity)
top_20_per_group = df2.groupby('race/ethnicity')['writing score'].nlargest(20)
# Extract the indices
top_20_indices = top_20_per_group.index
# Print the indices
print("Indices of the top 20 students with highest writing scores for each group:")
print(top_20_indices.tolist())
```

```python
# Task 3: Thống kê số lượng học sinh của group A theo parental level of education
groupA_parental_level_of_education_students = df2.groupby(df2['race/ethnicity'] == 'group A').value_counts()
print(groupA_parental_level_of_education_students)
```

Base on the answer we can conclude that the relation between education levels with average scores of children is yes. The higher educational levels of parents the higher average score of children

```python
# Task 4: Có phải xu hướng chung, cha mẹ có học vấn càng cao thì điểm số trung bình 3 môn của con cái cũng càng cao không? 
df2['avg_score'] = (df2['math score'] + df2['reading score'] + df2['writing score']) / 3
groups_data = df2.groupby(['parental level of education'])['avg_score'].mean()
ans = groups_data.sort_values(ascending=False)
print(ans)
```

If we use normal statistics we can not conclude the answer for the question whether the lunch affected to average score of students. We should use Hypothesis Testing right here, with critical values is 0.05, H0 is lunch not affected to avg score of students

```python
# Task 5: Chất lượng bữa ăn có tương phản với điểm số trung bình 3 môn của hs không? 
groups_lunch_data = df2.groupby(['lunch'])['avg_score'].mean()
ans_task_5 = groups_lunch_data.sort_values(ascending=False)
print(ans_task_5)
```

```python
# Scores for each lunch type
standard_scores = df2[df2['lunch'] == 'standard']['avg_score']
reduced_scores = df2[df2['lunch'] == 'free/reduced']['avg_score']
# Perform a z-test
z_stat, p_value = ztest(standard_scores, reduced_scores)
print(f"Z-statistic: {z_stat}, P-value: {p_value}")
if p_value < 0.05:
    print("Lunch is not affected to average score of students")
else:
    print("Lunch is affected to average score of students")
```

```python
# Task 6: Tìm ra top 10 học sinh có điểm toán cao nhất của mỗi group 
group_students = df2.groupby(['race/ethnicity'])['math score'].nlargest(10)
print(group_students)
```

The answer for task 7 is yes, the avg score of students who completed test preparation course is higher than students not completed it, with the same group,level of education of parent,lunch. This can explained by the effort of the students for training.

```python
# Task 7: Với những học sinh ở cùng 1 group, cùng parental level of education, cùng lunch, việc tham gia test preapration course có giúp học sinh đó có điểm trung bình 3 môn cao hơn những học sinh khác không? 
group_data_task7 = df2.groupby(['race/ethnicity','parental level of education','lunch','test preparation course'])['avg_score'].mean()
print(group_data_task7)
```
